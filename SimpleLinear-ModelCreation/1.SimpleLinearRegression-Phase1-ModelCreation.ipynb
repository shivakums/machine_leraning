```python
# panda is alibrary to handle file in table like structre or db related files and function of this panda library will be saved in this memory
# created the alias for panda as pd in code we will use pd instead of panda for shortcut
import pandas as pd
```

```python
#The read_csv is a method or function of pandas which will read the specified file and store the data in variable dataset
# As part of Modle Creation and learning we have done the data collection and preprocessing as this data is preprocessed data
dataset = pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
# store the input of the table in varaible, independent is input data so store the tab YearsExperience of dataset to independent
# As part of Modle Creation and learning procedure the sp%lit of input and output this step split the input split
independent = dataset[["YearsExperience"]]
independent
```

```python
# store the output of the table in varaible, dependent is output data so store the tab Salary of dataset to independent
# As part of Modle Creation and learning procedure the split of input and output this step split the output split
dependent = dataset[["Salary"]]
dependent
```

```python
# This sklearn is a library and has class model_selection and tran_test_split is a method or function specifically used to split the train and test data split
# As part of Modle Creation and learning procedure the split of train and test data is done
# it generate 4 different data, from input data it generate the train and test date and from out[ut data it generate train and test data 
#here x_ is input data and y_ is output data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0 )
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# the regressor.coef_ is function used to calculate the slope regressor is the model that we created and coef_ is function that calculates the weight/slope
# weight is a slope in which pattern the data varies
weight = regressor.coef_
weight
```

```python
# the regressor.intercept_ is function used to calculate the intercept. regressor is the model that we created and intercept_ is function that calculates the intercepts
# intercept_ is a function o
bias = regressor.intercept_
bias
```

```python
# regressor.predict is the functiom used to predict the output based on the x_test data and y_test is the actual data
#use to cross check the prediction
y_pred = regressor.predict(x_test)
```

```python
# sklearn.metric is library and r2.score is function used to create the metrics which will calculate the accuracy of the prediction 
# with the y_test which is actual test data and y_pred is the predicted data
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```

```python
# r_score is close to learn then the model is good
r_score
```

```python
# import pickle is a library to save the model
# provide file name in which the model can be saved the model the extention of file should ve sav for pickle format
import pickle
filename = "finalized_model_linear.sav"

```

```python
#pickle.dum() is a method to save the created model into the specific file in our local system
pickle.dump(regressor, open(filename, 'wb'))
```

```python
# how to load the saved model the pickle.load() is used to opne the model with file name mentioned in to the loaded_model
#the result variable is used to store the prediction with input mention into the parameter
loaded_model = pickle.load(open("finalized_model_linear.sav", 'rb'))
result=loaded_model.predict([[5]])
```
